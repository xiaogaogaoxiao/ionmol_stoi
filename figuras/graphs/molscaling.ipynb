```python
from scipy import interpolate
import numpy as np

def interp_XS(XY,ix,iy):
    x=XY[ix].values
    y=XY[iy].values
    n=len(x)
    tck = interpolate.splrep(x, y, s=0)
    xnew = np.arange(x[0], x[n-1], 0.1)
    ynew = interpolate.splev(xnew, tck, der=0)
    XYnew = pd.DataFrame({ix:xnew, iy:ynew})
    return XYnew
```
