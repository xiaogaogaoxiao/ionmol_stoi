# Make cubic spline interpolation on data

```python
from scipy import interpolate

def interp_XS(XY,ix,iy,kind):
    x=XY[ix].values
    y=XY[iy].values
    n=len(x)
    spline = interpolate.interp1d(x, y, kind=kind)
    xnew = np.arange(x[0], x[n-1], 0.1)
    ynew = spline(xnew)
    XYnew = pd.DataFrame({ix:xnew, iy:ynew})
    return XYnew
```
